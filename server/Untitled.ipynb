```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('./data/data.csv')
```

```python
finalList = []
for index,j in enumerate(data["genres"]):
    j = j[1:len(j)-1]
    
    mydict = {}
    if(len(j)>0):
        l = []
        for g in (j.split(",")):
            l.append(g[1:len(g)-1])
        mydict[index]=l
        finalList.append(mydict.copy())
    else:
        mydict[index]=[]
        finalList.append(mydict.copy())
    
print(finalList)
```

```python
pd.DataFrame.from_records(finalList)
```
